```python
essays = pd.read_csv("essays.csv")
essays.loc[essays['cEXT'] == 'n', 'cEXT'] = 0
essays.loc[essays['cEXT'] == 'y', 'cEXT'] = 1

essays.loc[essays['cNEU'] == 'n', 'cNEU'] = 0
essays.loc[essays['cNEU'] == 'y', 'cNEU'] = 1

essays.loc[essays['cAGR'] == 'n', 'cAGR'] = 0
essays.loc[essays['cAGR'] == 'y', 'cAGR'] = 1

essays.loc[essays['cCON'] == 'n', 'cCON'] = 0
essays.loc[essays['cCON'] == 'y', 'cCON'] = 1

essays.loc[essays['cOPN'] == 'n', 'cOPN'] = 0
essays.loc[essays['cOPN'] == 'y', 'cOPN'] = 1

essays
```

```python
import gensim


corpus = essays['TEXT'].to_list()

def read_corpus(corpus, tokens_only=False):
    for index, document in enumerate(corpus):
        preprocessed_document = gensim.utils.simple_preprocess(document)
        corpus[index] = preprocessed_document

        if tokens_only:
            yield preprocessed_document
        else:
            # For training data, add tags
            yield gensim.models.doc2vec.TaggedDocument(preprocessed_document, [index])

train_corpus = list(read_corpus(corpus[:2000]))
train_corpus_tokens_only = list(read_corpus(corpus[:2000], tokens_only=True))
test_corpus = list(read_corpus(corpus[2000:], tokens_only=True))
```

```python
model = gensim.models.doc2vec.Doc2Vec(vector_size=50, min_count=2, epochs=40)

model.build_vocab(train_corpus)

model.train(
    train_corpus, 
    total_examples=model.corpus_count, 
    epochs=model.epochs
)
```

```python
X_train = np.array([model.infer_vector(train_corpus_tokens_only[0])])

for document in train_corpus_tokens_only[1:]:
    row = model.infer_vector(document)
    X_train = np.append(X_train, [row], axis=0)
```

```python
y_train = essays['cNEU'][:2000].to_numpy()
y_train = y_train.astype('int')
```

```python
X_test = np.array([model.infer_vector(test_corpus[0])])

for document in test_corpus[1:]:
    row = model.infer_vector(document)
    X_test = np.append(X_test, [row], axis=0)
```

```python
y_test = essays['cNEU'][2000:].to_numpy()
y_test = y_test.astype('int')
```

```python
from sklearn.linear_model import LogisticRegression


logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_test_pred = logreg.predict_proba(X_test)
logreg.score(X_test, y_test)
```
